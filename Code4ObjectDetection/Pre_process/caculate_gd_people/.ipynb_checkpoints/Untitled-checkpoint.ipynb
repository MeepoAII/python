```python
# sum picture 1 total number of people (annotations)

k = 0
with open(in_path, 'r') as f:
    data = json.load(f)
    for i in range(len(data["annotations"])):
        if data["annotations"][i]['image_id'] == 1:
            k += 1
            
print(k)
```

```python
# sum picture 1 total number of people (prediction)

k = 0
with open(prediction_path, 'r') as f:
    data = json.load(f)
    for i in range(len(data)):
        if data[i]['image_id'] == 1:
            k += 1
            
print(k)
```

```python
# merge

with open(in_path, 'r') as f, open(prediction_path, 'r') as g:
    data_anno = json.load(f)
    data_pred = json.load(g)
    number_of_images = len(data_anno['images'])
    accuracy_list = []
    for i in range(number_of_images):
        a, b = 0, 0
        for j in range(len(data_anno['annotations'])):
            if data_anno['annotations'][j]['image_id'] == i:
                a += 1
        for j in range(len(data_pred)):
            if data_pred[j]['image_id'] == i:
                b += 1
        this_accuracy = 1-abs(a-b)/a
        accuracy_list.append(this_accuracy)
    final_average_accuracy = \
    sum(accuracy_list) / len(accuracy_list) * 100
    print("Final average accuracy = {:.2f}%".format(final_average_accuracy))
    print("This length of final average accuracy is ", len(accuracy_list))
        
        
    
    
        
        
    
    















```
